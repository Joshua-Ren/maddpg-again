```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

agent0_100_MADDPG = pd.read_csv("/Users/hongzhengnan/Downloads/MADDPG_100_0.csv")['Value']
agent3_100_MADDPG = pd.read_csv("/Users/hongzhengnan/Downloads/MADDPG_100_3.csv")['Value']
agent0_100_DDPG = pd.read_csv("/Users/hongzhengnan/Downloads/DDPG_100_0.csv")['Value']
agent3_100_DDPG = pd.read_csv("/Users/hongzhengnan/Downloads/DDPG_100_3.csv")['Value']


def smoothing(x,smoothingWeight):
    last=0
    smooth = []
    for i in (x):
        smoothed = last*smoothingWeight + (1 - smoothingWeight) * i
        smooth.append(smoothed)

        last = smoothed
    return smooth
```

```python
smooth_agent0_100_MADDPG = np.array(smoothing(agent0_100_MADDPG,0.95))
smooth_agent0_100_DDPG = np.array(smoothing(agent0_100_DDPG,0.95))
smooth_agent3_100_MADDPG = np.array(smoothing(agent3_100_MADDPG,0.95))
smooth_agent3_100_DDPG = np.array(smoothing(agent3_100_DDPG,0.95))
```

```python
# Generate some example data to plot
x = np.linspace(0,25000,1000)
y1 = agent0_100_MADDPG
y2 = smooth_agent0_100_MADDPG
y3 = agent0_100_DDPG
y4 = smooth_agent0_100_DDPG

fig_size = (15, 8)
# Set figure size in inches (width, height)
fig = plt.figure(figsize=fig_size)
# # Create a new figure object
ax = fig.add_subplot(1, 1, 1)
# # Add a single axes to the figure
# # Plot lines giving each a label for the legend and setting line width to 2
ax.plot(x, y1, linewidth=2, label='agent0_100_MADDPG',alpha=0.2)
ax.plot(x, y2, linewidth=2, label='smooth_agent0_100_MADDPG',color='blue')
ax.plot(x, y3, linewidth=2, label='agent0_100_DDPG',alpha=0.2)
ax.plot(x, y4, linewidth=2, label='smooth_agent0_100_DDPG',color='yellow')


# # Set the axes labels. Can use LaTeX in labels within $...$ delimiters.
ax.set_xlabel('$episode$', fontsize=15)
ax.set_ylabel('$reward$', fontsize=15)
ax.set_title('agent0_reward_along_episode', fontsize=15)
ax.grid('on')
# # Turn axes grid on
ax.legend(loc='upper right', fontsize=15)
# # Add a legend
fig.tight_layout()
# This minimises whitespace around the axes.
# fig.savefig("compare_max_avg_acc_error.pdf")
# Save figure to current directory in PDF format
```
