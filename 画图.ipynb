```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

noisy_SNR_0_MADDPG_agent0 = pd.read_csv("/Users/hongzhengnan/Downloads/noisy_SNR_0_MADDPG_agent0.csv")['Value']
noisy_SNR_0_MADDPG_agent3 = pd.read_csv("/Users/hongzhengnan/Downloads/noisy_SNR_0_MADDPG_agent3.csv")['Value']
noisy_SNR_20_MADDPG_agent0 = pd.read_csv("/Users/hongzhengnan/Downloads/noisy_SNR_20_MADDPG_agent0.csv")['Value']
noisy_SNR_20_MADDPG_agent3 = pd.read_csv("/Users/hongzhengnan/Downloads/noisy_SNR_20_MADDPG_agent3.csv")['Value']
noisy_SNR_minus20_MADDPG_agent0 = pd.read_csv("/Users/hongzhengnan/Downloads/noisy_SNR_minus20_MADDPG_agent0.csv")['Value']
noisy_SNR_minus20_MADDPG_agent3 = pd.read_csv("/Users/hongzhengnan/Downloads/noisy_SNR_minus20_MADDPG_agent3.csv")['Value']
epiode_len_25_MADDPG_agent0 = pd.read_csv("/Users/hongzhengnan/Downloads/epiode_len_25_MADDPG_agent0.csv")['Value']
epiode_len_25_MADDPG_agent3 = pd.read_csv("/Users/hongzhengnan/Downloads/epiode_len_25_MADDPG_agent3.csv")['Value']


agent0_noisy_SNR_0_MADDPG = pd.read_csv("/Users/hongzhengnan/Downloads/agent0_noisy_SNR_0_MADDPG.csv")['Value']
agent0_noisy_SNR_20_MADDPG = pd.read_csv("/Users/hongzhengnan/Downloads/agent0_noisy_SNR_20_MADDPG.csv")['Value']
agent0_noisy_SNR_minus20_MADDPG = pd.read_csv("/Users/hongzhengnan/Downloads/agent0_noisy_SNR_minus20_MADDPG.csv")['Value']
agent0_noisy_SNR_minus10_MADDPG = pd.read_csv("/Users/hongzhengnan/Downloads/agent0_noisy_SNR_minus10_MADDPG.csv")['Value']
agent0_noisy_SNR_minus15_MADDPG = pd.read_csv("/Users/hongzhengnan/Downloads/agent0_noisy_SNR_minus15_MADDPG.csv")['Value']



DDPG_default = pd.read_csv("/Users/hongzhengnan/Downloads/DDPG_default.csv")['Value']
MA_50_F = pd.read_csv("/Users/hongzhengnan/Downloads/MA_50_F.csv")['Value']
MA_minus5_F = pd.read_csv("/Users/hongzhengnan/Downloads/MA_minus5_F.csv")['Value']
MA_minus20_T = pd.read_csv("/Users/hongzhengnan/Downloads/MA_minus20_T.csv")['Value']
MA_minus20_F = pd.read_csv("/Users/hongzhengnan/Downloads/MA_minus20_F.csv")['Value']





def smoothing(x,smoothingWeight):
    last= -5
    smooth = []
    for i in (x):
        smoothed = last*smoothingWeight + (1 - smoothingWeight) * i
        smooth.append(smoothed)

        last = smoothed
    return smooth
```

```python
smooth_noisy_SNR_0_MADDPG_agent0 = np.array(smoothing(noisy_SNR_0_MADDPG_agent0,0.95))
smooth_noisy_SNR_0_MADDPG_agent3 = np.array(smoothing(noisy_SNR_0_MADDPG_agent3,0.95))
smooth_noisy_SNR_20_MADDPG_agent0 = np.array(smoothing(noisy_SNR_20_MADDPG_agent0,0.95))
smooth_noisy_SNR_20_MADDPG_agent3 = np.array(smoothing(noisy_SNR_20_MADDPG_agent3,0.95))
smooth_noisy_SNR_minus20_MADDPG_agent0 = np.array(smoothing(noisy_SNR_minus20_MADDPG_agent0,0.95))
smooth_noisy_SNR_minus20_MADDPG_agent3 = np.array(smoothing(noisy_SNR_minus20_MADDPG_agent3,0.95))
smooth_epiode_len_25_MADDPG_agent0 = np.array(smoothing(epiode_len_25_MADDPG_agent0,0.95))
smooth_epiode_len_25_MADDPG_agent3 = np.array(smoothing(epiode_len_25_MADDPG_agent3,0.95))

smooth_agent0_noisy_SNR_minus20_MADDPG = np.array(smoothing(agent0_noisy_SNR_minus20_MADDPG,0.95))
smooth_agent0_noisy_SNR_20_MADDPG = np.array(smoothing(agent0_noisy_SNR_20_MADDPG,0.95))
smooth_agent0_noisy_SNR_0_MADDPG = np.array(smoothing(agent0_noisy_SNR_0_MADDPG,0.95))
smooth_agent0_noisy_SNR_minus10_MADDPG = np.array(smoothing(agent0_noisy_SNR_minus10_MADDPG,0.95))
smooth_agent0_noisy_SNR_minus15_MADDPG = np.array(smoothing(agent0_noisy_SNR_minus15_MADDPG,0.95))

smooth_DDPG_default = np.array(smoothing(DDPG_default,0.95))
smooth_MA_50_F = np.array(smoothing(MA_50_F,0.95))
smooth_MA_minus5_F = np.array(smoothing(MA_minus5_F,0.95))
smooth_MA_minus20_T = np.array(smoothing(MA_minus20_T,0.95))
smooth_MA_minus20_F = np.array(smoothing(MA_minus20_F,0.95))

```

```python
# Generate some example data to plot
x = np.linspace(0,25000,1000)
y1 = noisy_SNR_0_MADDPG_agent0
y2 = smooth_noisy_SNR_0_MADDPG_agent0
y3 = noisy_SNR_20_MADDPG_agent0
y4 = smooth_noisy_SNR_20_MADDPG_agent0
y5 = noisy_SNR_minus20_MADDPG_agent0
y6 = smooth_noisy_SNR_minus20_MADDPG_agent0
y7 = epiode_len_25_MADDPG_agent0
y8 = smooth_epiode_len_25_MADDPG_agent0


fig_size = (15, 8)
# Set figure size in inches (width, height)
fig = plt.figure(figsize=fig_size)
# # Create a new figure object
ax = fig.add_subplot(1, 1, 1)
# # Add a single axes to the figure
# # Plot lines giving each a label for the legend and setting line width to 2
ax.plot(x, y1, linewidth=2, label='noisy_SNR_0_MADDPG_agent0',alpha=0.2)
ax.plot(x, y2, linewidth=2, label='smooth_noisy_SNR_0_MADDPG_agent0',color='blue')
ax.plot(x, y3, linewidth=2, label='noisy_SNR_20_MADDPG_agent0',alpha=0.2)
ax.plot(x, y4, linewidth=2, label='smooth_noisy_SNR_20_MADDPG_agent0',color='yellow')
ax.plot(x, y5, linewidth=2, label='noisy_SNR_minus20_MADDPG_agent0',alpha=0.2)
ax.plot(x, y6, linewidth=2, label='smooth_noisy_SNR_minus20_MADDPG_agent0',color = 'green')
ax.plot(x, y7, linewidth=2, label='epiode_len_25_MADDPG_agent0',alpha=0.2)
ax.plot(x, y8, linewidth=2, label='smooth_epiode_len_25_MADDPG_agent0',color = 'red')


# # Set the axes labels. Can use LaTeX in labels within $...$ delimiters.
ax.set_xlabel('$episode$', fontsize=15)
ax.set_ylabel('$mean_reward$', fontsize=15)
ax.set_title('agent0_reward_along_episode', fontsize=15)
ax.grid('on')
# # Turn axes grid on
ax.legend(loc='upper right', fontsize=15)
# # Add a legend
fig.tight_layout()
# This minimises whitespace around the axes.
# fig.savefig("compare_max_avg_acc_error.pdf")
# Save figure to current directory in PDF format
```
